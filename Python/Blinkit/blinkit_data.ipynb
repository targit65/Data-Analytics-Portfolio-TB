```python
#### **BUSINESS ANALYSIS**
```

```python
#Total Sales
total_sales = df['Sales'].sum()
#Average Sales
avg_sales = df['Sales'].mean()
#Average Ratings
avg_ratings = df['Rating'].mean()

# Display
print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.0f}")
print(f"Average Ratings: ${avg_ratings:,.0f}")
```

```python
#Graphs and Charts
```

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum() 

```

```python
plt.pie(sales_by_fat, labels=sales_by_fat.index,
       autopct='%.1f%%',
       startangle=90)
plt.title('Item Fat Content')
plt.axis('equal')
plt.show()
```

```python
ax = sns.countplot(x = 'Item Fat Content', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_by_outlet = df.groupby('Outlet Size')['Sales'].sum()
```

```python
plt.pie(sales_by_outlet, labels = sales_by_outlet.index,
        autopct = '%.1f%%',
        startangle = 90)

plt.title('Sales By Outlet')
plt.axis('equal')
plt.show()
```

```python
salesbyitem = df.groupby(['Item Type'], as_index=False)['Sales'].sum().sort_values(by ='Sales', ascending=False)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(x = 'Item Type', y = 'Sales', data = salesbyitem)
```

```python
sales_by_item = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)


fig = plt.figure(figsize=(12,6), frameon=False, dpi=100, constrained_layout=True); fig.set_constrained_layout_pads(w_pad=0, h_pad=0)
bars = plt.bar(sales_by_item.index, sales_by_item.values)

plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales By Item Type')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize=8)
    
   
plt.show()
```

```python
## Fat Content By Outlet for Total Sales
```

```python
fatCont_by_Outlet= df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
fatCont_by_Outlet = fatCont_by_Outlet[['Regular','Low Fat']]

ax = fatCont_by_Outlet.plot(kind ='bar', figsize = (8,5), title = 'Outlet Tier By Item Fat Content')
plt.xlabel('Outlet Location Type')
plt.ylabel('Sales')
plt.legend(title = 'Item Fat Content')
plt.tight_layout()
plt.show()
                            
```

```python
print('Total Sales By Outlet Estalishment Year')
```

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize= (10,6), frameon=False, dpi=100)
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-')

plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Outlet Establishment')

for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=8)
    
plt.tight_layout() 
plt.show()
```
